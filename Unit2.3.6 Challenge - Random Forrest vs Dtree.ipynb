```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import ensemble
from sklearn.model_selection import cross_val_score
from itertools import combinations
from sklearn import tree
%matplotlib inline
```

```python
df = pd.read_csv('/Users/joshuarkarnad/Desktop/Thinkful/Data/insurance.csv')
```

```python
features = df.drop('charges', 1)
X = features
X = pd.get_dummies(X)
len(X.columns)
```

```python
def dtreescoremaker(x, y, num_features, num_depth):
    dtree = tree.DecisionTreeRegressor(
        max_features= num_features,
        max_depth= num_depth,
        random_state = 1337
    )
    dtree.fit(x, y)
    dtreescore = cross_val_score(dtree, x, df.charges, cv=10)
    print('Max Features:', num_features)
    print('Max Depth:', num_depth)
    print('MSE Mean: ', dtreescore.mean())
    print('MSE Std:', dtreescore.std())
```

```python
for maxfeat in range(10):
    for maxdepth in range(12):
        dtreescoremaker(X, df.charges, (maxfeat + 1), (maxdepth + 2))
```

```python
def model_evaluator(dataframe, model_name, target, number_features):
    features = dataframe
    Y = target
    combo_list = combo_maker(features.columns, number_features)
    for combo in combo_list:
        combo = list(combo)
        X = pd.get_dummies(dataframe[combo])
        scores = cross_val_score(model_name, X, Y, cv=10)
        print(scores.mean(), scores.std())
        print(combo)
```

```python
def combo_maker(column_names, combo_size):
    return list(combinations(column_names, combo_size))
```

```python
rfc = ensemble.RandomForestRegressor()
```

```python
for i in range(10):
    model_evaluator(X, rfc, df.charges, (i + 1))
```
