```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
raw_data = pd.read_csv('/Users/joshuarkarnad/Desktop/Thinkful/Data/epicurious-recipes-with-rating-and-nutrition/epi_r.csv')
```

```python
X = raw_data.drop(['rating', 'title', 'calories', 'protein', 'fat', 'sodium'], 1)
Y = raw_data.rating
```

```python
xsmall = X.iloc[:, 1:10]
```

```python
Ysimple = (Y > Y.mean()).astype(float)
```

```python
Ysimple.head()
```

```python
Y.head()
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(xsmall,Ysimple)
```

```python
svc.score(xsmall, Ysimple)
```
