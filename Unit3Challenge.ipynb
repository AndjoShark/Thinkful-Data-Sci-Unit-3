```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import ensemble
from sklearn.model_selection import cross_val_score
from sklearn import linear_model
from sklearn import preprocessing
%matplotlib inline
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import RidgeClassifier
from sklearn.linear_model import Lasso
```

```python
df = pd.read_csv('/Users/joshuarkarnad/Desktop/Thinkful/Data/insurance.csv')
df['agebybmi'] = df['age'] / df['bmi']
df['childrenbyage'] = df['children'] / df['age']
df['overweight'] = (df['bmi'] > 25).astype(int)
df['overweightsmoker'] = (df['overweight'] & df['smoker'] == 1).astype(int)
df['big charge'] = (df['charges'] > 25000).astype(int)
```

```python
df = df.drop('charges', axis = 1)
df = pd.get_dummies(df)
```

```python
df.head()
```

```python
trainsize = int(df.shape[0] / 5)
df_train = df.iloc[trainsize:, :].copy()
df_test = df.iloc[:trainsize, :].copy()
```

```python
X_train = df_train.loc[:, ~(df_train.columns).isin(['big charge'])]
Y_train = df_train['big charge']
```

```python
lr = linear_model.LogisticRegression(penalty='l1')
```

```python
lr.fit(X_train, Y_train)
```

```python
lr.score(X_train, Y_train)
## No difference in l1 or l2
```

```python
for alpharidge in range(1, 200, 1):
    ridger = linear_model.RidgeClassifier(alpha=alpharidge)
    ridger.fit(X_train, Y_train)
    best_ridge = [0, 0]
    ridge_score = ridger.score(X_train, Y_train)
    if ridge_score > best_ridge[1]:
        best_ridge = [alpharidge, ridge_score]
        print(best_ridge)
```

```python
## Best score is [84, 0.95144724556489257]
```

```python
lass = linear_model.Lasso(alpha=2)
lass.fit(X_train, Y_train)
lass_score = lass.score(X_train, Y_train)
print(lass_score)
```

Best Model is Ridge Logistic Classifier because it has the highest accuracy but it is not too much different from the other two. This makes sense because there aren't many features involved in this data set so overfitting is not much of a problem.
